## Indexing and Slicing

```python
#import necessary libraries
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
df = pd.DataFrame(boston.data, columns=[boston.feature_names])
df['medv'] = pd.Series(boston.target)
```

```python
df.head()
```

```python
# select rows 0,1,2 (but not 3)
df[0:3]
```

```python
# select the last row in the df
df[-1:]
```

```python
# select the first 3 rows (rows 0,1,2)
df[:3]
# df[3:]
## df[1:]
```

To create a fresh copy of the df DataFrame we use the syntax y = x.copy() just writing y=x references to same dataframe

#### Slicing Subsets of Rows and Columns in Python

```python
df.loc[0:2, ['CRIM', 'AGE', 'RAD']]
# df.loc[ (df['AGE']>70), ['CRIM', 'AGE', 'RAD']]
# df.loc[0:2, 'CRIM':]
```

```python
df.iloc[0:3, 5:]
```
