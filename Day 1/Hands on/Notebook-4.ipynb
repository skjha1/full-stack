###Why numpy when we have python DS?

```
#python array
%timeit list(range(1000000))
%memit list(range(1000000))
```

```
#numpy array
%timeit np.arange(1000000, dtype='int64')
%memit np.arange(1000000, dtype='int64')
```

 Numpy is faster than python native lists and also is memory efficient.

```
#interaction between numpy and python native list
%timeit np.asarray(range(1000000))
%memit np.asarray(range(1000000))
```

When numpy and python interact with each other, we get the worst timings and most consumption of memory. 
 
So, do less interaction between python datastructures and numpy.
