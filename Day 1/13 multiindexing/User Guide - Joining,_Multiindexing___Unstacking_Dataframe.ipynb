```
"""
  Lets start with a assumption that there are two classes of 300 and 500 students respectively and
  they have given there one fav subject a rating in range of 1-10 from 6 unique subject .
"""
n_studs = 300
HouseOne = pd.DataFrame({
    "Name":["Name_"+str(i) for i in range(n_studs)],
    "Subject":np.random.choice(["Subject_"+str(i) for i in range(6)], size=n_studs),
    "Rating":np.random.uniform(low=1, high=10, size=n_studs)
})

n_studs = 500
HouseTwo = pd.DataFrame({
    "Name":["Name_"+str(i) for i in range(n_studs)],
    "Subject":np.random.choice(["Subject_"+str(i) for i in range(6)], size=n_studs),
    "Rating":np.random.uniform(low=1, high=10, size=n_studs)
})
```

```
HouseOne.head()
```

```
HouseTwo.head()
```

Let's say for some reason we want to compare average subject ratings from both the Houses, how do we do that?<br />
Let's first get the average rating for each subject given by first House.

```
subject_lookup_house_One = HouseOne.groupby("Subject")['Rating'].agg({"mean"})
subject_lookup_house_One
```

```
# renaming column name of mean to house_one rating cuz mean is not intuituve :/, mean of what ?
subject_lookup_house_One.rename(columns={"mean":"mean_rating_house_one"}, inplace=True) 
subject_lookup_house_One
```

```
# Similarly we can get for House Two 
subject_lookup_house_Two = HouseTwo.groupby("Subject")['Rating'].agg({"mean"})
subject_lookup_house_Two
```

```
subject_lookup_house_Two.columns = ['mean_rating_house_two']
subject_lookup_house_Two
```

Ok, now we have average from each house but still they are on two diffrent dataframe, so let's get them into one new dataframe by merging them.

```
"""
    Before merging let's think about it how would we mannually merge the look_ups we created above?,
    first we need to make sure there is something common between both lookups where we match the values,
    here Subject is common between both the lookups right?!, but if you notice subject is as index of the lookup, 
    luckily pandas provides a functionality to merge based on index values too. 

"""
merged_avgs = pd.merge(subject_lookup_house_One, subject_lookup_house_Two, left_index=True, right_index=True)
another_approch_for_merged_avgs = subject_lookup_house_One.merge(subject_lookup_house_Two, left_index=True, right_index=True) # this is another approach will give you same result
merged_avgs
```

```
print(another_approch_for_merged_avgs.equals(merged_avgs))
another_approch_for_merged_avgs
```

That's noice, but what if we want to reflect these Mean values in the original Dataframe accross each row where a particular subject is present.?<br /><br />

First let's think of the steps on our own, we have a lookup for mean value of each subject and we want to make new column in original dataset with mean values reflecting to there corresponding subject, so values common between original data & the lookup is subject

```
subject_lookup_house_One # Here subject is on Index
```

```
HouseOne.head() # Here subject is on column
```

```
HouseOne = HouseOne.merge(subject_lookup_house_One, how='left', left_on="Subject", right_index=True)
HouseOne
```

##### Another Approach

```
# We can reset_ the index column and merge both dataframe and lookup on column, in reality you will not always get the lookup values on where common value is on index, so let's see another approach
## Resetting the index column
subject_lookup_house_Two = subject_lookup_house_Two.reset_index()
subject_lookup_house_Two
```

```
## Now in our second scenerio we have the Subject in our lookup as a column so to merge column
HouseTwo = pd.merge(HouseTwo, subject_lookup_house_Two, how='left', on="Subject") # i suggest to go through the documentation of merge function to see the parameters
HouseTwo
```

## Concatenating

  Now we have two DataFrames for each house let's say if we want to make it into one single dataframe we can do that by appending one of the dataframe into Other rowwise or columnwise

```
houses_appended_row_wise = pd.concat([HouseOne, HouseTwo], axis=0) # say with me axis=0 is rowwise & axis=1 is columnwise
houses_appended_row_wise
```

```
houses_appended_column_wise = pd.concat([HouseOne, HouseTwo], axis=1)
houses_appended_column_wise
```

#Note 
Notice the shape of new Dataframe is 500 and there are some NaN in rows that's because HouseOne had 300 rows and HouseTwo had 500 rows, so for extra 200 rows in HouseTwo the HouseOne has no values so it is filled with NaN, same is true for Rowwise if there is some extra column while appending

### MultiIndexing

```
rating_to_grade = {
    10 : "O",
     9 : "E",
     8 : "A",
     7 : "B",
     6 : "C",
     5 : "D",
     4 : "E",
     3 : "F",
     2 : "Nailed it.!",
     1 : "Legendary"
}

HouseOne['discrete_rating'] = HouseOne['Rating'].apply(np.ceil)
HouseOne['Grade'] = HouseOne['discrete_rating'].map(rating_to_grade)
HouseOne
```

```
"""
    Let's say you want to do analysis on combination of two categorical columns, you can simply pass the columns in a list inside groupby function
"""
combination_of_subject_grade = HouseOne.groupby(["Subject", "Grade"])["Name"].count().to_frame()
combination_of_subject_grade # here you can see the combinations subject and grade and the count of combination of subject and grade and there respective counts
```

# Note
Notice the index, now subject & Grade are on index level, you can access any particular row by simply passing tupple 

```
combination_of_subject_grade.loc[("Subject_1", "A")]
```

## Unstacking

The groupby look_up we created above isn't very intuitive and it becomes hard to track values if size of one of the object increases, so we can use unstack function to get a table format.

```
combination_of_subject_grade.unstack()
```
