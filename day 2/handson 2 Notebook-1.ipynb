```python
data = [['Alex',10,'01-01-2011'],['Bob',12,'07-01-2008'],['Clarke',13,'01-31-2007'],
        ['Alex',10,'01-01-2011'],['David',12,'06-02-2008'],['Clarke',13,'05-01-2007']]
data = pd.DataFrame(data,columns=['Name','Age','Birthday'])
data
```

```python
#Converting to Datetime Format
data['Birthday'] = pd.to_datetime(data['Birthday'])
data['Birth_Year'] = data['Birthday'].dt.year
data['Birth_Month'] = data['Birthday'].dt.month
data
```

```python
#Finding & Removing duplicate records
data.duplicated().sum()
data.drop_duplicates(keep='first')
```

```python
#Here duplicated returns true if there is a repition in the values present in name only .
data.drop_duplicates(subset='Name',keep='first')
```
