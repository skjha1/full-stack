### Now lets remove datetime ambiguity

```python
train_data_test=train.copy(deep=True)
```

```python
def modify_date(row):
    row = str(row)
    year = row[:4]
    month = row[4:6]
    day = row[6:8]
    return f"{day}/{month}/{year}"

train_data_test['due_in_date']= [modify_date(row[6]) for row in train.iloc[:,:].values]
```

```python
for col in ['clear_date','posting_date','document_create_date','document_create_date_1','due_in_date']:
    train_data_test[col]= pd.to_datetime(train_data_test[col])
```
